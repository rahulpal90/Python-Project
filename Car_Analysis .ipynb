```python
# Selling price and prsent Price distribution 
plt.figure(figsize=(10,6))
sns.histplot(car['Selling_Price'] , kde = True , label = "Selling Price")
sns.histplot(car['Present_Price'] , kde = True , label = " Present Price")
plt.title('Selling price and prsent Price distribution')
plt.legend()
plt.show()
```

```python
# distribution kilometers Driven

plt.figure(figsize= (10,6))
sns.histplot(car['Kms_Driven'],bins= 30 ,kde = True ,color = 'b')
plt.title('Distribution of Kilometers Driven')
plt.show()
```

```python
# pair plot 
sns.pairplot(car)
plt.suptitle('Pairplot of Numerical Columns', y=1.02)
plt.show()
             
```

```python
numerical_data = car.select_dtypes(include=['float64','int64'])
plt.figure(figsize=(10,6))
sns.heatmap(numerical_data.corr() ,annot = True)
plt.title('Correlection Heatmap')
plt.show()
```
